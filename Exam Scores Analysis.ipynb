```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#Drop Unamed Column
df = df.drop('Unnamed: 0', axis = 1)
```

```python
print(df)
```

```python
plt.Figure(figsize = (5,5))
ax = sns.countplot(data = df, x = 'Gender')
ax.bar_label(ax.containers[0])
```

```python
# From above graph we have understood that the quatity of female and male is almost equl but female is slightlly more then male
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean','ReadingScore': 'mean','WritingScore': 'mean'})
print(gb)
```

```python
eh = sns.heatmap(data = gb, annot = True)
plt.title("Realationship between Student's Scores and ParentEduc" )
plt.show()
```

```python
#From Above chart we have concluded that parents those who have higher education and degree their students (child) socred relativlly more than students of parents those who don't
```

```python
gb3 = df.groupby('WklyStudyHours').agg({'MathScore': 'mean','ReadingScore': 'mean','WritingScore': 'mean'})
eh = sns.heatmap(data = gb3, annot = True)
plt.title("Realationship between Student's Scores and WklyStudyHours")
plt.show()
```

```python
#Studets those who studied for more than 10 hours scored more marks than others it shows students studied for longer hours scored better marks than those who studied for less hours
```

```python
gb2 = df.groupby('PracticeSport').agg({'MathScore': 'mean','ReadingScore': 'mean','WritingScore': 'mean'})
eh = sns.heatmap(data = gb2, annot = True)
plt.title("Realationship between Student's Scores and PracticeSport")
plt.show()
```

```python
#As above  chart is showing that PracticeSport have slight impact on studets marks
```

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean','ReadingScore': 'mean','WritingScore': 'mean'})
eh = sns.heatmap(data = gb1, annot = True)
plt.title("Realationship between Student's Scores and ParentMaritalStatus")
plt.show()
```

```python
#From above chart we have concluded that ParentMaritalStatus have negligible or slightlly impact on students Scores
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
print(groupA)
```

```python
l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mlist, labels = l, autopct = "%1.2f%%") 
plt.title('Distribution of EthnicGroup')
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'EthnicGroup')
ax.bar_label(ax.containers[0])
plt.title('Count of stundest from EthnicGroup')
plt.show()
```

```python
## Key Findings:
##1 Students who studied for more than 10 hours per week and whose parents had higher education consistently scored higher across subjects.
##2 The Parent's Marital Status showed negligible impact on students' scores.
##3 Engaging in sports practice showed a slight positive correlation with academic performance.
##4 A distribution analysis of Ethnic Groups (Group A to Group E) revealed that Group C had the highest proportion of students (31.99%), followed by Group D (26.05%).
```
